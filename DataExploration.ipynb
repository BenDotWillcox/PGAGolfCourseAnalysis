```python
from sklearn.neighbors import NearestNeighbors

def labels_ranked_by_proximity(input_label, X, labels, n_neighbors=5):
    # Find the index of the input label
    input_label_index = labels.index(input_label)
    
    # Use NearestNeighbors to find the nearest neighbors of the input label
    nn = NearestNeighbors(n_neighbors=n_neighbors + 1)  # +1 to exclude the input label itself
    nn.fit(X)
    distances, indices = nn.kneighbors([X[input_label_index]])
    
    # Get the labels of the nearest neighbors
    ranked_labels = [(labels[idx], distances[0][i+1]) for i, idx in enumerate(indices[0][1:])]
    
    return ranked_labels

# Example usage
```

```python
input_label = 'Albany GC'  # Input your label here
labels_list = df_concat['course'].tolist()  # Convert Series to list
ranked_labels = labels_ranked_by_proximity(input_label, X_pca, labels_list, n_neighbors=10)
print(f"Labels ranked by proximity to '{input_label}':")
for label, distance in ranked_labels:
    print(f"Label: {label}, Distance: {distance}")
```

```python
def plot_kmeans_results_with_labels(X, labels, k, random_state=42):
    # Fit KMeans clustering with increased number of iterations and fixed random state
    kmeans = KMeans(n_clusters=k, n_init=1000, random_state=random_state)
    kmeans.fit(X)
    
    # Get cluster labels
    kmeans_labels = kmeans.labels_
    
    # Create a dictionary to store clusters and labels
    clusters = {f'Cluster {i+1}': [] for i in range(k)}
    
    # Add labels to clusters
    for i, label in enumerate(labels):
        clusters[f'Cluster {kmeans_labels[i]+1}'].append(label)
    
    # Plot clusters
    scatter = go.Scatter(
        x=X[:, 0],
        y=X[:, 1],
        mode='markers',
        text=labels,
        hoverinfo='text',
        marker=dict(color=kmeans_labels, size=8, opacity=0.7)
    )

    centroid = go.Scatter(
        x=kmeans.cluster_centers_[:, 0],
        y=kmeans.cluster_centers_[:, 1],
        mode='markers',
        marker=dict(color='black', size=12, symbol='x')
    )
    
    layout = go.Layout(
        plot_bgcolor='white'  # Set background color to white
    )

    fig = go.Figure(data=[scatter, centroid], layout=layout)
    fig.show()
    
    return clusters
```

```python
clusters = plot_kmeans_results_with_labels(X_pca, labels, k=20)
print(clusters)
```
